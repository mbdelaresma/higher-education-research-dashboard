```python
rp_df = df[['Research Title', 'Keywords', 'Abstract', 'SDG', 'NEDA Priority Area', 'Allocated Budget (PH Pesos) 1,000,000.00']].copy()
rp_df.columns = ['title', 'keywords', 'abstract', 'sdg', 'neda', 'budget']
```

```python
def remove_punctuations(data):
    punct = set(string.punctuation)
    punct.add("â€™")
    data = ("").join([ch for ch in data if ch not in punct])
    return data

def remove_stopwords(data):
    stop_words = stopwords.words('english')
    data = (" ").join([word for word in data.split() if word not in stop_words])
    return data

def remove_univ_words(data):
    univ_words = df[['University (Full Name)', 'University (Abbreviation)', 'Campus', 'City/Municipality', 'Province', 'Region']]
    univ_words = univ_words.values.tolist()
    univ_words = set(list(chain.from_iterable(univ_words)))
    univ_words.add("del")
    univ_words.add("norte")
    univ_words.add("college")
    univ_words.add("university")    
    univ_words.add("state") 
    univ_words.add("del norte")
    univ_words.add("del norte state")
    univ_words.add("norte state")
    univ_words.add("state college")
    univ_words.add("state university")
    univ_words.add("study conducted")
    data = (" ").join([word for word in data.split() if word not in univ_words])
    return data

def lemma_corpus(data):
    lemmatizer = WordNetLemmatizer()
    data = (" ").join([lemmatizer.lemmatize(word) for word in data.split()])
    return data

def replace_delimiter(data):
    data = re.sub(r"; ", " ", data)
    data = re.sub(r", ", " ", data)
    data = re.sub(r";", " ", data)
    data = re.sub(r",", " ", data)
    return data

def plot_loghist(x, bins):
    hist, bins = np.histogram(x, bins=bins)
    logbins = np.logspace(np.log10(bins[0]),np.log10(bins[-1]),len(bins))
    plt.hist(x, bins=logbins)
    plt.xscale('log')
    plt.title('Distribution (Log Scale)')
```

```python
# process titles
rp_df['title_processed'] = None
for idx in rp_df.index:
    try:
        data = rp_df['title'].loc[idx]
        data = remove_punctuations(data)
        data = remove_stopwords(data)
        data = remove_univ_words(data)
        data = lemma_corpus(data)
        rp_df['title_processed'].loc[idx] = data
    except:
        continue
        
# process keywords
rp_df['keywords_processed'] = None
for idx in rp_df.index:
    try:
        data = rp_df['keywords'].loc[idx]
        data = replace_delimiter(data)
        data = remove_stopwords(data)
        data = remove_univ_words(data)
        data = lemma_corpus(data)
        rp_df['keywords_processed'].loc[idx] = data
    except:
        continue
        
# process abstract
rp_df['abstract_processed'] = None
for idx in rp_df.index:
    try:
        data = rp_df['abstract'].loc[idx]
        data = remove_punctuations(data)
        data = remove_stopwords(data)
        data = remove_univ_words(data)
        data = lemma_corpus(data)
        rp_df['abstract_processed'].loc[idx] = data
    except:
        continue
```

```python
rp_df['keywords_processed'].fillna("", inplace=True)
rp_df['abstract_processed'].fillna("", inplace=True)

# create corpus
rp_df['corpus'] = None
for idx in rp_df.index:
    try:
        rp_df['corpus'].loc[idx] = (rp_df['title'].loc[idx] + 
                                    rp_df['abstract'].loc[idx] + 
                                    rp_df['keywords'].loc[idx])
    except:
        continue
        
# create processed corpus
rp_df['corpus_processed'] = None
for idx in rp_df.index:
    try:
        rp_df['corpus_processed'].loc[idx] = (rp_df['title_processed'].loc[idx] + 
                                              rp_df['abstract_processed'].loc[idx] + 
                                              rp_df['keywords_processed'].loc[idx])
    except:
        continue
```

```python
# sdg processessing
sdg_pred = rp_df[rp_df['sdg'].isnull()].index
sdg_idx = rp_df[(rp_df['sdg'].notnull())].index
sdg_df = rp_df[['corpus_processed', 'sdg']].loc[sdg_idx].copy()

print("Before Processing")
display(sdg_df['sdg'].apply(lambda x: type(x)).value_counts())
sdg_df['sdg_type'] = sdg_df['sdg'].apply(lambda x: type(x))

for idx in sdg_df[sdg_df['sdg_type'] == str].index:
    sdg_df['sdg'].loc[idx] = np.random.choice(sdg_df['sdg'].loc[idx].split(';'))
    
sdg_df = sdg_df.drop('sdg_type', axis=1)
sdg_df['sdg'] = sdg_df['sdg'].apply(lambda x: int(x))

print("After Processing")
display(sdg_df['sdg'].apply(lambda x: type(x)).value_counts())

sdg_df['sdg'].value_counts().plot(kind='bar');
```

```python
train, test = train_test_split(sdg_idx, test_size=0.33)

train_features = sdg_df['corpus_processed'].loc[train].values
test_features = sdg_df['corpus_processed'].loc[test].values
pred_features = rp_df['corpus_processed'].loc[sdg_pred].values

train_labels = sdg_df['sdg'].loc[train]
test_labels = sdg_df['sdg'].loc[test]

sbert = SentenceTransformer('allenai/specter')
sbert.max_seq_length =  512

train_embeddings = sbert.encode(train_features, convert_to_tensor=True)
test_embeddings = sbert.encode(test_features, convert_to_tensor=True)
pred_embeddings = sbert.encode(pred_features, convert_to_tensor=True)

train_embeddings = train_embeddings.cpu()
test_embeddings = test_embeddings.cpu()
pred_embeddings = pred_embeddings.cpu()
```

```python
fin_lr = LogisticRegression(C=0.01, max_iter=100, penalty='l2', solver='saga')
fin_lr = fin_lr.fit(train_embeddings, train_labels)
fin_preds = fin_lr.predict(pred_embeddings)
print(accuracy_score(fin_lr.predict(test_embeddings), test_labels))
```

```python
for idx1, idx2 in enumerate(sdg_pred):
    df['SDG'].loc[idx2] = fin_preds[idx1]
```

```python
kw_model = KeyBERT('allenai/specter')
def gen_kp(text):
    corpus = []
    corpus.append(text)
    tfidf_vectorizer = KeyphraseTfidfVectorizer()
    tfidf_vectorizer.fit_transform(corpus).toarray()
    keyphrases = kw_model.extract_keywords(docs=corpus, vectorizer=tfidf_vectorizer)
    keyphrases = (", ").join([phrase[0] for phrase in keyphrases[0]])
    return keyphrases
```

```python
%%capture
# generate keywords
rp_df['keyphrases_corpus_processed'] = None
for idx in rp_df.index:
    try:
        rp_df['keyphrases_corpus_processed'].loc[idx] = gen_kp(rp_df['corpus_processed'].loc[idx])
    
    except:
        continue
```
