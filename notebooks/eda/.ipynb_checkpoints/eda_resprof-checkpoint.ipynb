```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('research_profile_v1.xlsx')
df.iloc[:, 6:].head()
```

```python
df.columns
```

```python
df.loc[:, 'Region'].value_counts().plot(kind='bar')
plt.show()
```

```python
df['Allocated Budget (PH Pesos) 1,000,000.00'].loc[657] = [50000, 70000]
df['Allocated Budget (PH Pesos) 1,000,000.00'].loc[740] = 75000
df['Allocated Budget (PH Pesos) 1,000,000.00'].loc[844] = 422745

temp = df.loc[:, ['Region', 'Allocated Budget (PH Pesos) 1,000,000.00']].dropna()
temp = temp[~(temp['Allocated Budget (PH Pesos) 1,000,000.00'] == 0)]
temp = temp.explode('Allocated Budget (PH Pesos) 1,000,000.00')
temp = temp.groupby('Region')['Allocated Budget (PH Pesos) 1,000,000.00'].sum()
temp.plot(kind='bar')
```

```python
temp1 = df.loc[:, ['Region', 'Allocated Budget (PH Pesos) 1,000,000.00']].dropna()
temp1 = temp1[~(temp1['Allocated Budget (PH Pesos) 1,000,000.00'] == 0)]
temp1 = temp1.explode('Allocated Budget (PH Pesos) 1,000,000.00').groupby('Region')['Allocated Budget (PH Pesos) 1,000,000.00'].size()
```

```python
temp2 = pd.concat([temp, temp1], axis=1)
temp2.columns = ['Allocated Budget (PH Pesos) 1,000,000.00', 'Research Title']
temp2['Budget per Research'] = temp2['Allocated Budget (PH Pesos) 1,000,000.00']/temp2['Research Title']
temp2['Budget per Research'].sort_values(ascending=False).plot(kind='bar')
```

```python
df['Year (YYYY)'].notnull().value_counts().plot(kind='bar')
```

```python
df.iloc[:, 6].value_counts().sort_index().iloc[-6:-1].plot(kind='line')
plt.axhline(df.iloc[:, 6].value_counts().sort_index().iloc[-6:-1].mean(),
            c='red', linestyle='--', label='Mean Frequency')
plt.xticks(np.arange(2016, 2023, 1.0))
plt.yticks(np.arange(0, 400, 50))
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.legend(loc='lower right')
plt.show()
```

```python
test = (df.groupby(['Year (YYYY)', 'University (Full Name)'])['Research Title']
 .count().reset_index().set_index('Year (YYYY)')
 .pivot(columns='University (Full Name)').fillna(0).iloc[10:15]
)
test.columns = [j for i,j in test.columns]
test.iloc[:, [0, 6, 10, 19, 32, 37, 45, 51, 59]].plot(kind='line', figsize=(20,8))
plt.legend()
plt.show()
```

```python
(df.groupby(['Year (YYYY)', 'University (Full Name)'])['Research Title']
 .count().reset_index().set_index('Year (YYYY)')
 .pivot(columns='University (Full Name)').fillna(0).iloc[10:15]
 .plot(kind='bar', stacked=True, figsize=(20, 20))
)
```

```python
temp1 = df['University (Abbreviation)'].value_counts()
cmap = plt.get_cmap('Pastel1')
colors = cmap(np.linspace(0, 1, len(temp1[temp1 > 5])))
temp = df.groupby(['University (Abbreviation)', 'Campus'])['Research Title'].count().reset_index()
temp = temp[temp['University (Abbreviation)'].isin(temp1[temp1 > 5].index)]
univ_colors = dict()
for i, v in enumerate(temp1[temp1 > 5].index):
    univ_colors[v] = colors[i]
temp['colors'] = temp['University (Abbreviation)'].map(univ_colors)
# temp['Campus1'] = temp.iloc[:, 0] + '_Main'
# temp['Campus'] = np.where(temp['Campus'] == 'Main', temp['Campus1'], temp['Campus'])
temp['Campus'] = temp['University (Abbreviation)'] + '_' + temp['Campus']
# temp = temp.drop('Campus1', axis=1)

plt.figure(figsize=(20,10))
plt.bar(temp['Campus'], temp['Research Title'], color=temp['colors'], label='')
plt.xticks(rotation=90)
# plt.legend()
plt.show()
```
