```python
import pandas as pd
import numpy as np
import re

pd.set_option('display.max_rows', None)
```

```python
sample = pd.read_excel('research_profile_v1.xlsx')

print(sample.columns)

sample['Year (YYYY)'] = sample['Year (YYYY)'].astype('Int64')
sample['Year Published'] = sample['Year Published'].astype('Int64')
sample['Number of Citations as of Jan 2022'] = sample['Number of Citations as of Jan 2022'].astype('Int64')
sample['Region'] = sample['Region'].astype(str)


for i in range(7, 13):
    sample.iloc[:, i] = sample.iloc[:, i].str.strip()
    
sample.iloc[:, 6:].head()
```

```python
def format_filter_resprof(x):
    error_cols = []
  
    # YEAR
    # Year is integer
    if not isinstance(x[6], int):
        year_error = 'Year is not integer.'
        error_cols.append(year_error)
        
    if not isinstance(x[13], int):
        sdg_error = 'SDG is not integer.'
        error_cols.append(sdg_error) 
    
    if not isinstance(x[14], str):
        neda_error = 'NEDA Priority Area is not string.'
        error_cols.append(neda_error) 
    
    if not isinstance(x[15], str):
        journal_title_error = 'Journal Title is not string.'
        error_cols.append(journal_title_error) 
        
    if not isinstance(x[16], str):
        journal_volume_error = 'Journal Volume is not string.'
        error_cols.append(journal_volume_error) 
    
    if not isinstance(x[17], str):
        journal_issue_error = 'Journal Issue is not string.'
        error_cols.append(journal_issue_error) 
    
    if not isinstance(x[18], int):
        citations_error = 'Citations is not integer.'
        error_cols.append(citations_error) 
    
    
    if error_cols:
        return error_cols
    else:
        return 'No errors'
```
